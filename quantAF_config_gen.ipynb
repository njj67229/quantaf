```python
import panel as pn
css = '''
.bk.space-top {
    margin-top:6.5px;
}
.bk.space-top-check {
    margin-top:8.5px;
}
.bk.space-top-file {
    margin-top:4.5px;
}

.bk.space-left {
    margin-left:0px;
    margin-right:0px;
    padding-left:0px;
    padding-right:0px;
}
.bk.panel {
    background-color: #ffe5db;
}
'''
pn.extension(raw_css=[css])

ref_text_in = pn.widgets.TextInput(placeholder="Upload a file or provide a URL")
ref_file_in = pn.widgets.FileInput(css_classes=["space-top-file"])

fasta_text_in = pn.widgets.TextInput(placeholder="Upload a file or provide a URL")
fasta_file_in = pn.widgets.FileInput(css_classes=["space-top-file"])

gtf_text_in = pn.widgets.TextInput(placeholder="Upload a file or provide a URL")
gtf_file_in = pn.widgets.FileInput(css_classes=["space-top-file"])

read_len_in = pn.widgets.TextInput(value="90")

read_len_row = pn.Row(pn.pane.HTML("Read Length", css_classes=["space-top"]),pn.pane.HTML(width=0), read_len_in)
ref_row = pn.Row(pn.pane.HTML("Reference File", css_classes=["space-top"]),ref_text_in,ref_file_in)
fasta_row = pn.Row(pn.pane.HTML("fasta File", css_classes=["space-top"]), pn.layout.Spacer(width=23),fasta_text_in,fasta_file_in)
gtf_row = pn.Row(pn.pane.HTML("gtf File", css_classes=["space-top"]), pn.layout.Spacer(width=35),gtf_text_in,gtf_file_in)

add_check = pn.widgets.Checkbox(name="Additional Options")
additional_options_row = pn.Row(add_check)

file_pref_check = pn.widgets.Checkbox(name="Filename-prefix", css_classes=["space-top-check"])
file_pref_text = pn.widgets.TextInput(placeholder="Prefix", width=200, visible=False)
file_pref_row = pn.Row(file_pref_check,file_pref_text, width=200, height=35) 
file_pref_check.jslink(file_pref_text, value="visible")

trim_len_check = pn.widgets.Checkbox(name="Flank Trim Length",css_classes=["space-top-check"])
trim_len_text = pn.widgets.TextInput(placeholder="Trim Length", width=200, visible=False)
trim_row = pn.Row(trim_len_check,trim_len_text, width=200, height=35)
trim_len_check.jslink(trim_len_text, value="visible")

ext_spl_fst_check = pn.widgets.Checkbox(name="Extra Spliced Fasta",css_classes=["space-top-check"])
ext_spl_fst_text = pn.widgets.TextInput(placeholder="Upload file or provide URL", width=200, visible=False)
ext_spl_file_in = pn.widgets.FileInput(css_classes=["space-top-file"], visible=False)
ext_spl_row = pn.Row(ext_spl_fst_check, ext_spl_fst_text, ext_spl_file_in, width=350, height=35)
ext_spl_fst_check.jslink(ext_spl_file_in, value="visible")
ext_spl_fst_check.jslink(ext_spl_fst_text, value="visible")

ext_uspl_fst_check = pn.widgets.Checkbox(name="Extra Unspliced Fasta",css_classes=["space-top-check"])
ext_uspl_fst_text = pn.widgets.TextInput(placeholder="Upload file or provide URL", width=200, visible=False)
ext_uspl_file_in = pn.widgets.FileInput(css_classes=["space-top-file"],visible=False)
ext_uspl_row = pn.Row(ext_uspl_fst_check, ext_uspl_fst_text, ext_uspl_file_in, width=350, height=35)
ext_uspl_fst_check.jslink(ext_uspl_file_in, value="visible")
ext_uspl_fst_check.jslink(ext_uspl_fst_text, value="visible")

dedup_check = pn.widgets.Checkbox(name="Deduplicate Identical Sequences", height=25)
flank_lengths = pn.widgets.Checkbox(name="Consider Flank Lengths for Merge", height=25)

additional_opts_col = pn.Column(file_pref_row, trim_row,
                                ext_spl_row, ext_uspl_row, pn.pane.HTML(height=0),
                                dedup_check, flank_lengths, visible=False)
add_check.jslink(additional_opts_col,value="visible")


generate_config_button = pn.widgets.Button(name="Generate Config File!")
generated_text = pn.pane.Markdown("None", width = 150)

generate_config = '''
var text = ""
text += "trace.enabled=true;"
text += "trace.file='pipeline_trace.txt';"
text += "trace.fields = 'task_id,name,tag,status,exit,realtime,%cpu,%mem,rss,peak_rss,realtime';"
text += "params.pyroe.read_len=" + read_len.value + ";"
text += "params.pyroe.filename_pref="
if(file_pref_check.value) {
    text += file_pref_text.value + ";"
} else {
    text += '"";'
}
text += "extra_spliced"
generated_text.text = text
var element = document.createElement('a');
element.setAttribute('href', 'data:text/plain;charset=utf-8,' + encodeURIComponent(text));
element.setAttribute('download', "pyroe.txt");
element.style.display = 'none';
document.body.appendChild(element);
element.click();
document.body.removeChild(element);
'''

generate_config_button.js_on_click(args={"generated_text" : generated_text, "read_len" : read_len_in, 
                                         "file_pref_check" : file_pref_check, "file_pref_text" : file_pref_text,
                                         "ext_spl_fst_check": ext_spl_fst_check, "ext_spl_fst_text":ext_spl_fst_text,
                                         "ext_uspl_fst_check": ext_uspl_fst_check, "ext_uspl_fst_text":ext_uspl_fst_text},
                                         code=generate_config)


pyroe_col = pn.Column("# pyroe options",ref_row, fasta_row, gtf_row, read_len_row, additional_options_row, additional_opts_col, generate_config_button, generated_text, css_classes=["panel"], width=600
            )
pyroe_col
```
